# 3. Гистограммы/распределения признаков

Заменим diagnosis  - имеет на два занчения, 0 и 1. B(0) - доброкачественная, M (1) - злокачественная

```python
df_new = df.copy()
df_new['diagnosis'] = df_new['diagnosis'].replace('B',0)
df_new['diagnosis'] = df_new['diagnosis'].replace('M',1)
corr = df_new.corr()
corr
```

### Матрица корреляции

```python

plt.figure(figsize=(10, 8))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=90)
a=plt.yticks(range(len(corr)), corr.columns)
```

### Фильтрация признаков

```python
b_treshould = 0.2
t_treshould = 0.8
mask = ((corr >= b_treshould) & (corr <= t_treshould)) | ((corr <= -b_treshould) & (corr >=-t_treshould))
top_corr=corr[mask]
# top_vars
new_corr = corr[mask.any(axis=1)]
columns = list(corr.columns)
pairs_col1 = []
pairs_col2 = []
pairs_col3 = []

for col in columns:
    for c in columns:
        if c == col:
            break
        if (corr.at[c,col] >= b_treshould and corr.at[c,col] <= t_treshould) or (corr.at[c,col] <= -b_treshould and corr.at[c,col] >= -t_treshould):
#             print(corr.at[c,col])
            pairs_col1.append(col)
            pairs_col2.append(c)
            pairs_col3.append(corr.at[c,col])
#             pairs.append([col,c,corr.at[c,col]])
top_corr_features=pd.DataFrame({'f1':pairs_col1,'f2':pairs_col2,'corr':pairs_col3})
```

### Нормолизация данных

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
print(scaler.fit(df_new[top_features]))
```

```python
print(scaler.mean_)
```

```python
model_data = scaler.transform(df_new[top_features])
model_data
```

```python
model_data.shape, df_new['diagnosis'].values.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(model_data, df_new['diagnosis'].values, test_size=0.30, random_state=42)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_classifire = KNeighborsClassifier()
knn_classifire.fit(X_train, Y_train)
```

```python
Y_test_predict = knn_classifire.predict(X_test)
Y_test_predict
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score,precision_recall_curve
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
f1_score(Y_test, Y_test_predict, average='binary'), accuracy_score(Y_test, Y_test_predict), recall_score(Y_test, Y_test_predict,average='binary')
```

```python
cm = confusion_matrix(Y_test, Y_test_predict)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=knn_classifire.classes_)
disp.plot()
plt.show()
```

```python
Y_test_predict_prob = knn_classifire.predict_proba(X_test)
fpr, tpr, thresholds = roc_curve(Y_test, Y_test_predict_prob[:, 1])
roc_auc_score(Y_test, Y_test_predict_prob[:, 1])
```

```python
plt.plot(fpr, tpr)
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
```
