### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
A = np.zeros(10)
A[4] = 1
print(A)
A = A.reshape(2,5)
A
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
B = np.arange(10,50)
print('array B:\n',B)
B = np.flip(B,0)
print('reversed array B:\n',B)
even_B = B[1:40:2]
print('even B:\n',even_B)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
C = np.arange(0,9)
C = C.reshape(3,3)
print(C)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
D = np.random.random(4*3*2)
# print(D)
D = np.reshape(D,(4,3,2))
print(D)
max = np.max(D)
min = np.min(D)
print(f'max value: {max}, min value: {min}')
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
A1 = np.arange(24).reshape(6,4)
B1 = np.arange(12).reshape(4,3)
print(A1,'\n===============\n',B1)
C1 = A1@B1
C1
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
mu, sigma = -1, 5.0 # среднее сзначение и среднеквадратичное отклонени
A2 = np.random.normal(mu, sigma, size=(7, 7))
print(A2)
mean = np.average(A2)
std = np.std(A2)
print(mean,std)
B2 = (A2-mean)/std
print(B2)
```

### Найдите максимальное значение 'total_bill'

```python
max=df['total_bill'].max()
print(f'Максимальное значение колонки total_bill из датасета tips: {max}')
```

### Найдите количество курящих людей

```python
smoke_p = df[df['smoker']== 'Yes'].shape[0]
print(f'Количество курящих людей: {smoke_p}')
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
days = df['day'].unique()
for d in days:
    mean_bill = df[df['day']==d]['total_bill'].mean()
    print(f'День: {d}.\nСредний bill: {mean_bill}\n')
# print(f'Количество курящих людей: {days}')
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
median_bill = df['total_bill'].median()
print('median_bill: ',median_bill)
median_filter = df['total_bill'] > median_bill
df_median_bill = df[median_filter]
df_median_bill
```

```python
df_median_bill['total_bill'].hist(bins=20)
```

```python
sex = df['sex'].unique()
for s in sex:
    mean_bill = df[df['sex']==s]['total_bill'].mean()
    print(f'Пол: {s}.\nСредний bill: {mean_bill}\n')
# print(f'Количество курящих людей: {days}')
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
df['smoker'] = df['smoker'].replace({'No':0,'Yes':1})
df
```
